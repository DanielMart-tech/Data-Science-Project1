```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import csv
```

```python
football_data = pd.read_csv("results.csv")
football_data
```

### away score de Argentina

```python
punt_arg_away = football_data[(football_data['away_team'] == "Argentina") & (football_data["away_score"] >=6)]
punt_arg = punt_arg_away[['date', 'away_score']]
punt_arg
```

```python
dic_arg = punt_arg.set_index('date').T.to_dict('list')
dic_arg
```

```python
lista_nueva_arg = [x[:4] for x in dic_arg.keys()]
lista_nueva_arg
```

```python
lista_nueva_arg1 = [x for x in dic_arg.values()]
lista_nueva_arg1
```

```python
list_pun_arg = []
for punt in lista_nueva_arg1:
    list_pun_arg.append(punt[0])
    
print(list_pun_arg)
```

```python
fig = plt.figure()
grafic_arg = plt.bar(lista_nueva_arg, list_pun_arg)
plt.xlabel('fechas')
plt.ylabel('puntajes')
plt.title('grafico away score Argentina')
fig.savefig('puntajesArgentina.png', dpi = fig.dpi)
```
